# **Prueba Naive**  
Objetivo: Realizar una primera aproximación de forecast para una serie sintética y poder realizar una predicción para el step siguiente

# Librerías

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

# Funciones

```python
def plot_series(time, series, format="-", start=0, end=None):
    """Permite pintar una serie según eje x (tiempo) e y (series) 
      con punto de inicio y fin"""
      
    plt.plot(time[start:end], series[start:end], format)
    plt.xlabel("Tiempo")
    plt.ylabel("Valor")
    plt.grid(True)
```

# Creación serie temporal sintética
En este notebook podemos ver paso a paso la construcción de la serie sintética https://colab.research.google.com/drive/1ovwzq8HovMnP8yEcnltTqZ8yL3JeFrQB#scrollTo=WFc-PIuTfe5B

```python
def trend(time, slope=0):
    """Creación de una linea recta como tendencia 
    por un tiempo determinado(time) y una pendiente(slope)"""
    return slope * time

def seasonal_pattern(season_time):
    """Patrón arbitrario estacional"""
    return np.where(season_time < 0.1,
                    np.cos(season_time * 7 * np.pi),
                    1 / np.exp(5 * season_time))

def seasonality(time, period, amplitude=1, phase=0):
    """Repite el mismo patrón cada cierto período (amplitude)"""
    season_time = ((time + phase) % period) / period
    return amplitude * seasonal_pattern(season_time)

def noise(time, noise_level=1, seed=None):
    """Ruido aleatorio"""
    rnd = np.random.RandomState(seed)
    return rnd.randn(len(time)) * noise_level

#Definición del tiempo de la serie => 10 años
time = np.arange(10 * 365 + 1, dtype="float32")

#Variables de la serie
baseline = 10     #base de inicio a 10
amplitude = 40    #se repite la estacionalidad cada 40 puntos
slope = 0.005     #pendiente de la tendencia
noise_level = 3   #nivel de ruido

# Creación de la serie
# 1.- Tendencia
series = trend(time, slope)
# 2.- Añadimos la estacionalidad
series += baseline + seasonality(time, period=365, amplitude=amplitude)
# 3.- Añadimos el ruido
series += noise(time, noise_level, seed=51)

#Pintamos la serie
plt.figure(figsize=(10, 6))
plot_series(time, series)
plt.show()
```

# Split datos

Realiza el split

```python
# 9 años de train y 1 de test
SPLIT_TIME = 3285

# datos train
time_train = time[:SPLIT_TIME]
x_train = series[:SPLIT_TIME]
plt.figure(figsize=(10, 6))
plot_series(time_train, x_train)
plt.title ('Datos train')
plt.show()

#datos_test
time_test = time[SPLIT_TIME:]
x_test = series[SPLIT_TIME:]
plt.figure(figsize=(10, 6))
plot_series(time_test, x_test)
plt.title ('Datos test')
plt.show()
```

# Método Naive de Forecast  
Predecir según valor del día anterior

```python
# datos a validar con predicción de método naive
# la predicción será el valor anterior
naive_forecast = series[SPLIT_TIME - 1:-1] 
```

```python
len(naive_forecast)
```

```python
#Comparamos en test las dos series
plt.figure(figsize=(10, 6))
plot_series(time_test, x_test)
plot_series(time_test, naive_forecast)
```

```python
#Zoom sobre los 150 primeros puntos para ver que el forecast es un lag de tiempo posterior
plt.figure(figsize=(10, 6))
plot_series(time_test, x_test, start=0, end=150)
plot_series(time_test, naive_forecast, start=1, end=151)
```

```python
#Cálculo del error
print('Error mse: ', mean_squared_error(x_test, naive_forecast))
print('Error mae: ', mean_absolute_error(x_test, naive_forecast))
```

# Método Moving Average  
Predecir según una media de una ventana anterior

```python
def moving_average_forecast(series, window_size):
  """Predicción basada en la media de los n valores anteriores (window_size)
     Para window_size=1, sería lo mismo que naive forecast"""
  
  forecast = []
  for time in range(len(series) - window_size):
    #añadimos a la lista la media de los n valores o window determinada
    forecast.append(series[time:time + window_size].mean())
  
  return np.array(forecast)  #devolvemos en formato array la lista de predicciones con la media de la ventana
```

### Primer approach

```python
WINDOW_SIZE = 30
moving_avg = moving_average_forecast(series, WINDOW_SIZE)
moving_avg = moving_avg[SPLIT_TIME - WINDOW_SIZE:]   #Nos quedamos con los datos de test

plt.figure(figsize=(10, 6))
plot_series(time_test, x_test)
plot_series(time_test, moving_avg)
```

```python
#Cálculo del error
print('Error mse: ', mean_squared_error(x_test, moving_avg))
print('Error mae: ', mean_absolute_error(x_test, moving_avg))
```

### Segundo approach
Ya que no aprende la tendencia ni estacionalidad, podemos aplicarle la diferencia sobre cada punto de su homólogo de anterior estación para quitarle tendencia y estacionalidad  
Aquí tenemos estacionalidad de 1 año

```python
diff_series = (series[365:] - series[:-365])  #desde el primer día del segundo año eliminanos su homologo -365
diff_time = time[365:]                        #tenemos 365 menos lags  de tiempo

plt.figure(figsize=(10, 6))
plot_series(diff_time, diff_series)
plt.show()
```

```python
WINDOW_SIZE = 50
diff_moving_avg = moving_average_forecast(diff_series, WINDOW_SIZE)
diff_moving_avg = diff_moving_avg[SPLIT_TIME - 365 - WINDOW_SIZE:]  #Nos quedamos con los datos de test

plt.figure(figsize=(10, 6))
plot_series(time_test, diff_series[SPLIT_TIME - 365:])
plot_series(time_test, diff_moving_avg)
plt.show()
```

Debemos añadirle al forecast realizado el valor de 365 días anterior para tener la misma referencia que x_test, que es realmente con quien estamos testeando

```python
diff_moving_avg_plus_past = series[SPLIT_TIME - 365:-365] + diff_moving_avg

plt.figure(figsize=(10, 6))
plot_series(time_test, x_test)
plot_series(time_test, diff_moving_avg_plus_past)
plt.show()
```

```python
#Cálculo del error
print('Error mse: ', mean_squared_error(x_test, diff_moving_avg_plus_past))
print('Error mae: ', mean_absolute_error(x_test, diff_moving_avg_plus_past))
```
