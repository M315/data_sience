```python
import pandas as pd
import sklearn as skl
import seaborn as sns
```

```python
red = pd.read_csv('data/winequality-red.csv', sep = ';', header = 0)
white = pd.read_csv('data/winequality-white.csv', sep = ';', header = 0)
```

```python
def show_df(df):
    print("Data Frame:\n\n", df.describe)
    print("\nShape: ",df.shape)
    print("\nNulls:\n", df.isnull().values.sum())
    print("\nTypes:")
    for column in df:
        print("\t", '{:<20}'.format(column), '{:>8}'.format(str(df[column].dtype)))
    
#show_df(red)
show_df(white)
```
