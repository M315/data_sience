## Outliers

Vamos a mirar si tenemos outliers.

```python
def estadisticas(df):
    print('{:<23}'.format('Column'), '{:<23}'.format('Mean'), '{:<23}'.format('Std'), '{:<10}'.format('Median'), '{:<10}'.format('Min'), '{:<10}'.format('Max'),'\n')
    for col in df:
        print('{:<23}'.format(col), '{:<23}'.format(df[col].mean()), '{:<23}'.format(df[col].std()), '{:<10}'.format(df[col].median()), '{:<10}'.format(df[col].min()), '{:<10}'.format(df[col].max()))
```

```python
estadisticas(white)
```

Podemos ver como en algunas columnas la desviacion estandar es casi tan grande como la media y que la diferencia entre el minimo y/o el maximo con la mediana es significativo.

Vamos a eliminar las filas que en alguna columna tengan un z-score mayor a 3 en valor absoluto.

```python
def clean_outliers(df):
    new = df.copy()
    for col in new:
        new = new[(np.abs(stats.zscore(new[col])) < 3)]
  
    return new
```

```python
w = clean_outliers(white)
estadisticas(w)
```

Podemos ver como ahora las estadisticas son mejores y ya no hay tantos datos extremos.

Hacemos lo mismo con el otro data frame.

```python
r = clean_outliers(red)
estadisticas(r)
```

## Correlacion + Eliminar columnas

A priori no hay ninguna columna que sea obvio que se deba eliminar. Pero vamos a mirar si hay alguna correlacion entre ellas que nos ayude a decidir si alguna se debe eliminar para que no afecte en el modelo.

```python
sns.heatmap(w.corr(), annot = True)
```

```python
sns.heatmap(r.corr(), annot = True)
```

La unica correlación relativamente alta es entre el azucar residual y la desidad en el vino blanco, cosa que tiene sentido, pero no es suficientemente alta como para eliminar una de las dos columnas.
