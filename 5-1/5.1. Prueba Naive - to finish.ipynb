# **Prueba Naive**  
Objetivo: Realizar una primera aproximación de forecast para una serie sintética y poder realizar una predicción para el step siguiente

# Librerías

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

# Funciones

```python
def plot_series(time, series, format="-", start=0, end=None):
    """Permite pintar una serie según eje x (tiempo) e y (series) 
      con punto de inicio y fin"""
      
    plt.plot(time[start:end], series[start:end], format)
    plt.xlabel("Tiempo")
    plt.ylabel("Valor")
    plt.grid(True)
```

# Creación serie temporal sintética
En este notebook podemos ver paso a paso la construcción de la serie sintética https://colab.research.google.com/drive/1ovwzq8HovMnP8yEcnltTqZ8yL3JeFrQB#scrollTo=WFc-PIuTfe5B

```python
def trend(time, slope=0):
    """Creación de una linea recta como tendencia 
    por un tiempo determinado(time) y una pendiente(slope)"""
    return slope * time

def seasonal_pattern(season_time):
    """Patrón arbitrario estacional"""
    return np.where(season_time < 0.1,
                    np.cos(season_time * 7 * np.pi),
                    1 / np.exp(5 * season_time))

def seasonality(time, period, amplitude=1, phase=0):
    """Repite el mismo patrón cada cierto período (amplitude)"""
    season_time = ((time + phase) % period) / period
    return amplitude * seasonal_pattern(season_time)

def noise(time, noise_level=1, seed=None):
    """Ruido aleatorio"""
    rnd = np.random.RandomState(seed)
    return rnd.randn(len(time)) * noise_level

#Definición del tiempo de la serie => 10 años
time = np.arange(10 * 365, dtype="float32")

#Variables de la serie
baseline = 10     #base de inicio a 10
amplitude = 40    #se repite la estacionalidad cada 40 puntos
slope = 0.005     #pendiente de la tendencia
noise_level = 3   #nivel de ruido

# Creación de la serie
# 1.- Tendencia
series = trend(time, slope)
# 2.- Añadimos la estacionalidad
series += baseline + seasonality(time, period=365, amplitude=amplitude)
# 3.- Añadimos el ruido
series += noise(time, noise_level, seed=51)

#Pintamos la serie
plt.figure(figsize=(10, 6))
plot_series(time, series)
plt.show()
```

# Split datos

Realiza el split

```python
INIT_TEST_TIME = 9 * 365

train = np.array(series[:INIT_TEST_TIME])
test = np.array(series[INIT_TEST_TIME:])

train_time = np.arange(INIT_TEST_TIME, dtype="float32")
test_time = np.arange(365, dtype="float32") + INIT_TEST_TIME

#Pintamos la serie
plt.figure(figsize=(10, 6))
plot_series(train_time, train)
plt.show()

#Pintamos la serie
plt.figure(figsize=(10, 6))
plot_series(test_time, test)
plt.show()
```

# Método Naive de Forecast  
Predecir según valor del día anterior

```python
naive = np.array(train[-365:])

#Comparamos en test las dos series
plt.figure(figsize=(10, 6))
plot_series(test_time, test)
plot_series(test_time, naive)
plt.show()
```

```python
def mae(test, pred):
    err = abs(test - pred)
    return err.sum() / len(err)

def mse(test, pred):
    err = abs(test - pred)
    err **= 2
    return err.sum() / len(err)
```

```python
#Cálculo del error
print("Error mse:\t", mse(test, naive))
print("Error mae:\t", mae(test, naive))
```

# Método Moving Average  
Predecir según una media de una ventana anterior

```python
def sma(train, k = 100):
    k = min(len(train), k)
    ma = np.zeros(k)
    
    cum_sum = 0
    for i in range(k):
        cum_sum += train[-(1 + i)]
    
    for i in range(k):
        ma[i] = cum_sum / k
        cum_sum -= train[-(1 + k - i)]
        cum_sum += ma[i]
    
    return ma
```

### Primer approach

```python
k = 365
ma = sma(train, k)
ma_time = np.arange(k, dtype="float32") + INIT_TEST_TIME

#Comparamos en test las dos series
plt.figure(figsize=(10, 6))
plot_series(ma_time, test[:k])
plot_series(ma_time, ma)
plt.show()
```

```python
#Cálculo del error
print("Error mse:\t", mse(test[:k], ma))
print("Error mae:\t", mae(test[:k], ma))
```

### Segundo approach
Ya que no aprende la tendencia ni estacionalidad, podemos aplicarle la diferencia sobre cada punto de su homólogo de anterior estación para quitarle tendencia y estacionalidad  
Aquí tenemos estacionalidad de 1 año

```python
def sma_2(train, k = 10, n = 365, season = 365):
    ma = np.zeros(season)
    k = min(k, len(train) // k)
    i = 0
    
    while len(train) >= season and i < k:
        i += 1
        ma += train[-season:]
        train = train[:-season]
    
    return ma[:n] / k
```

```python
days = 100
ma_2 = sma_2(train, k = 3, n = days)
```

```python
#Comparamos en test las dos series
plt.figure(figsize=(10, 6))
plot_series(test_time[:days], test[:days])
plot_series(test_time[:days], ma_2)
plt.show()
```

```python
#Cálculo del error
print("Error mse:\t", mse(test[:days], ma_2))
print("Error mae:\t", mae(test[:days], ma_2))
```
